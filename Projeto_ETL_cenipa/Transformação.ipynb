```python
import pandas as pnd
import pandera as pad

valores_ausentes = ['**','***','****','*****','###!','####','NULL']
data = pnd.read_csv('ocorrencias.csv',sep=';',parse_dates=['ocorrencia_dia'], dayfirst =True, na_values=valores_ausentes)#le o arquivo com o separador e ve uma das var como data
schema = pad.DataFrameSchema(
    columns={
        "codigo":pad.Column(pad.Int, required=False),
        "codigo_ocorrencia":pad.Column(pad.Int),
        "ocorrencia_uf": pad.Column(pad.String, pad.Check.str_length(2,2), nullable=True),
        "ocorrencia_dia": pad.Column(pad.DateTime, nullable = True),
        "ocorrencia_hora": pad.Column(pad.String,pad.Check.str_matches(r'^([0-1]?[0-9]|[2][0-3]):([0-5][0-9])(:[0-5][0-9])?$'),nullable = True)
    }
)#valida se os dados presentes estão verdades
try:
    schema.validate(data)#valida se esta tudo ok, se tiver dados errados retorna um erro
except BaseException as ex:
    print('Erro na formatação dos dados: {}'.format(ex))
else:
    print('Dados OK')
```

```python
# filtro ocorrencia classificação == acidente 
filtro = data.ocorrencia_classificacao == 'ACIDENTE'
data.loc[filtro,['codigo_ocorrencia','ocorrencia_cidade']]
```

```python
# filtro ocorrencia classificação == acidente e estado sp
filtro = data.ocorrencia_classificacao == 'ACIDENTE'
filtro2 = data.ocorrencia_uf == 'SP'
data.loc[filtro & filtro2,['codigo_ocorrencia','ocorrencia_cidade']]
```

```python
# filtro ocorrencia classificação == acidente ou incidente grave e estado sp
    #filtro = (data.ocorrencia_classificacao == 'ACIDENTE') | (data.ocorrencia_classificacao == 'INCIDENTE GRAVE')
filtro = data.ocorrencia_classificacao.isin(['ACIDENTE','INCIDENTE GRAVE'])   
filtro2 = data.ocorrencia_uf == 'SP'
data.loc[filtro & filtro2,['codigo_ocorrencia','ocorrencia_cidade']]
```

```python
# filtro ocorrencia cidade começa com a letra s
    #filtro = (data.ocorrencia_classificacao == 'ACIDENTE') | (data.ocorrencia_classificacao == 'INCIDENTE GRAVE')  
filtro = data.ocorrencia_cidade.str[0] == 'S'
data.loc[filtro,['codigo_ocorrencia','ocorrencia_cidade']]
```

```python
# filtro ocorrencia cidade termina com a letra s
    #filtro = (data.ocorrencia_classificacao == 'ACIDENTE') | (data.ocorrencia_classificacao == 'INCIDENTE GRAVE')  
filtro = data.ocorrencia_cidade.str[-1] == 'S'
data.loc[filtro,['codigo_ocorrencia','ocorrencia_cidade']]
```

```python
# filtro ocorrencia cidade termina com a letra as
    #filtro = (data.ocorrencia_classificacao == 'ACIDENTE') | (data.ocorrencia_classificacao == 'INCIDENTE GRAVE')  
filtro = data.ocorrencia_cidade.str[-2:] == 'AS'
data.loc[filtro,['codigo_ocorrencia','ocorrencia_cidade']]
```

```python
# filtro ocorrencia cidade contem com a letra as
    #filtro = (data.ocorrencia_classificacao == 'ACIDENTE') | (data.ocorrencia_classificacao == 'INCIDENTE GRAVE')  
filtro = data.ocorrencia_cidade.str.contains('AS')
data.loc[filtro,['codigo_ocorrencia','ocorrencia_cidade']]
```

```python
# filtro ocorrencia cidade contem com a letra as ou da
    #filtro = (data.ocorrencia_classificacao == 'ACIDENTE') | (data.ocorrencia_classificacao == 'INCIDENTE GRAVE')  
filtro = data.ocorrencia_cidade.str.contains('AS|DA')
data.loc[filtro,['codigo_ocorrencia','ocorrencia_cidade']]
```

```python
# filtro ocorrencia dia para o dia 21 de todos os anos e meses
    #filtro = (data.ocorrencia_classificacao == 'ACIDENTE') | (data.ocorrencia_classificacao == 'INCIDENTE GRAVE')  
filtro = data.ocorrencia_dia.dt.day == 21
data.loc[filtro,['codigo_ocorrencia','ocorrencia_cidade']]
```

```python
# filtro ocorrencia dia para o dia 21 de todos meses do dia 21 do ano de 2019
    #filtro = (data.ocorrencia_classificacao == 'ACIDENTE') | (data.ocorrencia_classificacao == 'INCIDENTE GRAVE')  
filtro = data.ocorrencia_dia.dt.day == 21
filtro2 = data.ocorrencia_dia.dt.year == 2019
data.loc[filtro&filtro2,['codigo_ocorrencia','ocorrencia_cidade']]
```

```python
# filtro ocorrencia dia para oS dias entre 21 e 24 
filtro = (data.ocorrencia_dia.dt.day > 21) & (data.ocorrencia_dia.dt.day < 24) 
data.loc[filtro,['codigo_ocorrencia','ocorrencia_cidade']]
```

```python
# nova coluna com data e hora
data['ocorrencia_dia_hora'] = pnd.to_datetime(data.ocorrencia_dia.astype(str) +' '+ data.ocorrencia_hora)
```
